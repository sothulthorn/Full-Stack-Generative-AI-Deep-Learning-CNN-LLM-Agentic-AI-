```python
# import
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img, save_img
```

```python
# load image from folder and convert to PIL object
cat_image = load_img('images/cat.jpg')

# print properties of image
print(cat_image.size)
print(type(cat_image))
print(cat_image.mode)
print(cat_image.format)
```

```python
# show image in viewer
cat_image.show()
```

```python
# show image inline
import matplotlib.pyplot as plt

plt.imshow(cat_image)
plt.show()
```

```python
# convert PIL image to numpy array
cat_image = img_to_array(cat_image)
```

```python
# printing the properties of the numpy array
print(type(cat_image))
print(cat_image.shape)
print(cat_image)
```

```python
# convert numpy array to PIL image
cat_image = array_to_img(cat_image)
```

```python
plt.imshow(cat_image)
plt.show()
```

```python
# load image from folder and convert to PIL object
cat_image_gray = load_img('images/cat.jpg', color_mode='grayscale')

# save image to folder
cat_image_gray = img_to_array(cat_image_gray)
save_img('images/cat_gray.jpg', cat_image_gray)
```

```python
# load image from folder and convert to PIL object
cat_image = load_img('images/cat.jpg', color_mode='grayscale')
cat_image = img_to_array(cat_image)
print(cat_image.shape)
```

```python
import numpy as np

# change channel order of the array
cat_image = np.rollaxis(cat_image, 2, 0)
print(cat_image.shape)
```
